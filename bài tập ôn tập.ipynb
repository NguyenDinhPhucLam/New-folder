```python
# sử dụng while hoặc for để tính tổng các số nguyên tử 1 đến n (n <= 1000)
n = int(input())
tong = 0
for i in range (1, n+1):
    tong += i
print(tong)
```

```python
# nhập vào hai số nguyên a b, nếu âm thì in ra yes, ngược lại thì no
a, b = map(int, input().split())
if a < 0 and b < 0:
    print("Yes")
else:
    print("No")
```

```python
#kiểm tra n có phải số nguyên tố
n = int(input("Nhập n: "))
if n < 2:
    print("Không phải số nguyên tố")
else:
    kt = True
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            kt = False
            break
    if kt:
        print("Số nguyên tố")
    else:
        print("Không phải số nguyên tố")
```
